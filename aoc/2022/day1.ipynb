```typescript
import Aoc from '../aoc.ts';

const aoc = await Aoc.load(2022, 1);
```

```typescript
const data = aoc.input.trim().split('\n\n').map((food) => food.split('\n').map((cals) => parseInt(cals, 10)));
data;
```

```typescript
const totals = data.map((elf) => elf.reduce((acc, val) => acc + val, 0)).sort((a, b) => b - a);
totals
```

```typescript
void await aoc.answer1(totals[0]);
```

```typescript
void await aoc.answer2(totals.slice(0, 3).reduce((acc, value) => acc + value, 0));
```
