```typescript
import Aoc from "../aoc.ts";

const aoc = await Aoc.load(2024, 6);
```

```typescript
let input = `....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#...`;
input = aoc.input;

type Map = string[][];
type Point = [y: number, x: number];

const parsed: Map = input.split("\n").map((line) => line.split(""));

const dirs: Point[] = [
  [-1, 0],
  [0, 1],
  [1, 0],
  [0, -1],
];

const w = parsed[0].length;
const h = parsed.length;
const startIndex = input.indexOf("^");
const sx = startIndex % (w + 1);
const sy = Math.floor(startIndex / (w + 1));
const sp: Point = [sy, sx];

function move(map: Map, cb: (point: Point, d: number) => boolean) {
  let d = 0;
  let [y, x] = sp;
  while (true) {
    const [dy, dx] = dirs[d];
    const [ny, nx] = [y + dy, x + dx];
    if (nx < 0 || nx >= w || ny < 0 || ny >= h) break;
    if (map[ny][nx] !== "#") {
      if (!cb([ny, nx], d)) break;
      x = nx;
      y = ny;
    } else {
      d = (d + 1) % 4;
    }
  }
}
```

```typescript
const visited = new Set<string>();
move(parsed, ([y, x]) => {
  visited.add([y, x].join(","));
  return true;
});

visited.size;
```

```typescript
await aoc.answer1(visited.size);
```

```typescript
function check(map: Map): boolean {
  const seen = new Set<string>();
  let looping = false;
  move(map, ([y, x], d) => {
    const key = [d, x, y].join(",");
    if (seen.has(key)) {
      looping = true;
      return false;
    }
    seen.add(key);
    return true;
  });
  return looping;
}

let loops = 0;
for (const p of visited) {
  const [y, x] = p.split(",").map(Number);
  const map = parsed.map((line) => line.slice());
  if (sx === x && sy === y) continue;
  map[y][x] = "#";
  if (check(map)) loops++;
}

loops;
```

```typescript
await aoc.answer2(loops);
```
