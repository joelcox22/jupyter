```typescript
import Aoc from "../aoc.ts";

const aoc = await Aoc.load(2024, 1);
```

```typescript
let input = `3   4
4   3
2   5
1   3
3   9
3   3`;
input = aoc.input;
input;
```

```typescript
const lists: number[][] = [[], []];
for (const line of input.split("\n")) {
  const [a, b] = line.split("   ").map(Number);
  lists[0].push(a);
  lists[1].push(b);
}

lists;
```

```typescript
const sorted = lists.map((list) => list.sort((a, b) => a - b));

sorted;
```

```typescript
const distances = lists[0].map((_, i) => Math.abs(sorted[1][i] - sorted[0][i]));

distances;
```

```typescript
const total = distances.reduce((acc, d) => acc + d, 0);

// total;
await aoc.answer1(total);
```

```typescript
const similarities = lists[0].map(
  (v) => lists[1].filter((v2) => v2 === v).length * v
);
similarities;
```

```typescript
const totalSimilarity = similarities.reduce((acc, s) => acc + s, 0);

totalSimilarity;

await aoc.answer2(totalSimilarity);
```
