```typescript
import Aoc from "../aoc.ts";

const aoc = await Aoc.load(2024, 2);
```

```typescript
let input = `7 6 4 2 1
1 2 7 8 9
9 7 6 2 1
1 3 2 4 5
8 6 4 4 1
1 3 6 7 9`;
input = aoc.input;

const parsed = input.split("\n").map((line) => line.split(" ").map(Number));
parsed
```

```typescript
const safe = parsed.map((line) => {
  const direction = line[1] > line[0];
  return line.map((n, i) => {
    if (i === 0) return n;
    const last = line[i - 1];
    if (Math.abs(n - last) < 1 || Math.abs(n - last) > 3) return 'wrong diff';
    if (i > 0 && (n > line[i-1]) !== direction) return 'invalid order';
    return n;
  });
});

safe;
```

```typescript
const safe2 = safe.map((line) => line.filter((v) => typeof v === 'number').length === line.length);

await aoc.answer1(safe2.filter(Boolean).length);
```

```typescript
const potentialLines = parsed.map((line) => {
  let newLines = [];
  for (let i=0; i<line.length; i++) {
    newLines.push([...line.slice(0, i), ...line.slice(i+1)]);
  }
  return newLines.map((line) => {
    const direction = line[1] > line[0];
    return line.map((n, i) => {
      if (i === 0) return n;
      const last = line[i - 1];
      if (Math.abs(n - last) < 1 || Math.abs(n - last) > 3) return 'wrong diff';
      if (i > 0 && (n > line[i-1]) !== direction) return 'invalid order';
      return n;
    });
  }).map((line) => line.filter((v) => typeof v === 'number').length === line.length).filter((v) => v).length > 0;
});

await aoc.answer2(potentialLines.filter((v) => v).length)
```
