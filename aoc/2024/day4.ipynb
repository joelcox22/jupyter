```typescript
import Aoc from "../aoc.ts";

const aoc = await Aoc.load(2024, 4);
```

```typescript
let input = `MMMSXXMASM
MSAMXMSMSA
AMXSXMAAMM
MSAMASMSMX
XMASAMXAMM
XXAMMXXAMA
SMSMSASXSS
SAXAMASAAA
MAMMMXMMMM
MXMXAXMASX`;
input = aoc.input;

const w = input.split("\n")[0].length;
const h = input.split('\n').length;

void 0
```

```typescript
const lines = input.split('\n').map((line) => line.trim());

let output: string[][] = new Array(h).fill('').map(() => new Array(w).fill('.'));

function find(x: number, y: number, dx: number, dy: number, match: string) {
  let result = true;
  for (let i=0, X=x, Y=y; i<match.length; X+=dx, Y+=dy, i++) {
    if (X < 0 || Y < 0 || X >= w || Y >= h) {
      result = false;
      break;
    }
    if (lines[Y].charAt(X) !== match.charAt(i)) {
      result = false;
      break;
    }
  }
  if (result) {
    for (let i=0, X=x, Y=y; i<match.length; X+=dx, Y+=dy, i++) {
      if (X < 0 || Y < 0 || X >= w || Y >= h) break;
      output[Y][X] = match[i];
    }
  }
  return result;
}

const directions = [
  [0, 1], // down
  [1, 0], // right
  [1, 1], // down right
  [1, -1], // up right
  [-1, 1], // down left
  [-1, -1], // up left
  [0, -1], // up
  [-1, 0] // left
];

let count = 0;
const result: string[] = [];

for (let y=0; y<h; y++) {
  result[y] = '';
  for (let x=0; x<w; x++) {
    let found = false;
    for (const [dx, dy] of directions) {
      if (find(x, y, dx, dy, 'XMAS')) {
        found = true;
        count++;
      }
    }
    // if (found) count++;
  }
}

console.log(`${output.map((l) => l.join('')).join('\n')}\n\n${count}`)
```

```typescript
await aoc.answer1(count);
```

```typescript
const W = w+1; // ignore newline
let count2 = 0;

output = new Array(h).fill('').map(() => new Array(w).fill('.'));

for (let x=0; x<w-2; x++) {
  for (let y=0; y<h-2; y++) {
    let block = input.substr(y*W+x, 3)+input.substr((y+1)*W+x, 3)+input.substr((y+2)*W+x, 3);
    
    if (block.match(/(M.M.A.S.S)|(M.S.A.M.S)|(S.S.A.M.M)|(S.M.A.S.M)/)) {
      output[y][x] = block[0];
      output[y][x+2] = block[2];
      output[y+1][x+1] = 'A';
      output[y+2][x] = block[6];
      output[y+2][x+2] = block[8];
      // console.log(input.substr(y*W+x, 3)+'\n'+input.substr((y+1)*W+x, 3)+'\n'+input.substr((y+2)*W+x, 3)+'\n----');
      count2++;
    }
  }
}

console.log(`${output.map((l) => l.join('')).join('\n')}\n\n${count2}`)
```
